```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.pib.gov.in/PressReleasePage.aspx?PRID=2044754"

headers = {
"User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "lxml")

tables = soup.find_all("table")

data = []

for table in tables:
    rows = table.find_all("tr")

    for row in rows[1:]:
        cols = row.find_all(["td","th"])
        cols = [c.get_text(strip=True) for c in cols]

        if len(cols) >= 2:
            data.append(cols)

df = pd.DataFrame(data)

print(df.head())

df.to_csv("../data/raw/seed_varieties_dataset.csv", index=False)

print("Saved: seed_varieties_dataset.csv")
```

```python
import pandas as pd

df = pd.read_csv("../data/processed/seed_varieties_cleaned.csv")

# forward fill category
df.iloc[:,0] = df.iloc[:,0].ffill()

# remove rows where crop missing
df = df[df.iloc[:,1].notna()]

# remove category only rows
df = df[df.iloc[:,1] != df.iloc[:,0]]

# rename columns properly
df.columns = [
"Category",
"Crop",
"Name of Variety",
"Variety/Hybrid",
"Sponsoring organization",
"State",
"Description",
"Extra"
]

# drop duplicates
df = df.drop_duplicates()

# reset index
df = df.reset_index(drop=True)

df.to_csv("../data/processed/seed_varieties_final.csv", index=False)

print(df.head())
print("Final dataset saved")
```

```python
state_col = "State"

# split multiple states
df[state_col] = df[state_col].astype(str)

df[state_col] = df[state_col].str.replace(" and ", ",")
df[state_col] = df[state_col].str.replace(";", ",")
df[state_col] = df[state_col].str.replace("/", ",")

# split into list
df[state_col] = df[state_col].str.split(",")

# explode rows
df = df.explode(state_col)

# clean whitespace
df[state_col] = df[state_col].str.strip()

# remove empty rows
df = df[df[state_col] != ""]
df = df[df[state_col] != "nan"]

# reset index
df = df.reset_index(drop=True)

# save dataset
df.to_csv("../data/processed/seed_varieties_state_split.csv", index=False)

print("Saved: seed_varieties_state_split.csv")
print(df.head())
```

```python
def extract_features(text):
    text = str(text).lower()
    
    return pd.Series({
        "drought": int("drought" in text),
        "saline": int("saline" in text),
        "rainfed": int("rainfed" in text),
        "irrigated": int("irrigated" in text),
        "disease_resistant": int("disease" in text),
        "early_maturity": int("early" in text),
        "high_yield": int("high yield" in text)
    })

features = df["Description"].apply(extract_features)

df = pd.concat([df, features], axis=1)
```

```python
df
```

```python
def recommend_seed(crop, state, drought=0, rainfed=0, saline=0):
    
    filtered = df[
        (df["Crop"].str.lower()==crop.lower()) &
        (df["State"].str.lower()==state.lower())
    ].copy()
    
    filtered["score"] = (
        filtered["drought"]*drought +
        filtered["rainfed"]*rainfed +
        filtered["saline"]*saline +
        filtered["high_yield"]
    )
    
    filtered = filtered.sort_values("score", ascending=False)
    
    return filtered[["Name of Variety","score"]].head(3)
```

```python
recommend_seed(
    crop="rice",
    state="maharashtra",
    drought=1,
    rainfed=1
)
```
